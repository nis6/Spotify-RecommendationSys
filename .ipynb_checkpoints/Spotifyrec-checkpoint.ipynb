**One Hot Encoder function**

```python
def ohe(df,columns,newcolname):
  ohe_features=pd.get_dummies(df[columns])
  feature_names=ohe_features.columns
  ohe_features.columns=[newcolname+'|'+str(i) for i in feature_names]
  ohe_features.reset_index(drop=True,inplace=True)
  return ohe_features
```

**Final Feature Set**

```python
def create_feature_set(df,float_cols):

  ###############                           TF-IDF FEATURE SET

  tfidf=TfidfVectorizer()
  tfidf_matrix=tfidf.fit_transform(df['merged_genre_list'].apply(lambda x: " ".join(x)))

  # df['merged_genre_list'].apply(lambda x: " ".join(x)) creates a list of documents where each list of genre col is taken as document
  # tfidf.fit_transform calculates importance of a genre to that doc[/list of genre] in the collection of genre lists 
  genre_df = pd.DataFrame(tfidf_matrix.toarray())

  # converts the matrix type object to usable nd array
  genre_df.columns = ['genre' + "|" + i for i in tfidf.get_feature_names()]
  genre_df.reset_index(drop=True, inplace=True)

  ###############                           ONE HOT ENCODING
  year_ohe = ohe(df, 'year','year') # *0.5 #?
  popularity_ohe = ohe(df, 'popularity_bins','pop')# * 0.15 #?

  ###############                         SCALING FLOAT COLUMNS
  floats=df[float_cols].reset_index(drop=True)
  feature_names=floats.columns
  scaler=MinMaxScaler()
  floats_scaled=pd.DataFrame(scaler.fit_transform(floats),columns=feature_names)

  ###############                           FINAL FEATURE-SET
  final=pd.concat([genre_df,floats_scaled,year_ohe,popularity_ohe],axis=1)

  #add song id
  final['id']=df['id'].values

  return final
```

# Without User Authentication

Client Credentials Flow

The Client Credentials flow is used in server-to-server authentication. Only endpoints that do not access user information can be accessed. The advantage here in comparison with requests to the Web API made without an access token, is that a higher rate limit is applied.

As opposed to the Authorization Code Flow, you will not need to set SPOTIPY_REDIRECT_URI, which means you will never be redirected to the sign in page in your browser.

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

sp_cred_object=SpotifyClientCredentials()
sp_dev = spotipy.Spotify(auth_manager=sp_cred_object) 
# only app authentication developer can use this client anytime

#if in case token is expired
token_info=sp_cred_object.get_access_token() # returns  dict with previously stored token info
if sp_cred_object.is_token_expired(token_info): 
    new_token_info=sp_cred_object.refresh_access_token(token_info['refresh_token']) #returns a dict with new access token
    sp_dev = spotipy.Spotify(auth=new_token_info['access_token'])
```

```python
def create_necessary_outputs(playlist_name,id_dic, df):
    """ 
    Pull songs from a specific playlist.

    Parameters: 
        playlist_name (str): name of the playlist you'd like to pull from the spotify API
        id_dic (dic): dictionary that maps playlist_name to playlist_id
        df (pandas dataframe): spotify datafram
        
    Returns: 
        playlist: all songs in the playlist THAT ARE AVAILABLE IN THE KAGGLE DATASET
    """
    
    #generate playlist dataframe
    playlist = pd.DataFrame()


    for ix, i in enumerate(sp_user.playlist(id_dic[playlist_name])['tracks']['items']):
        #print(i['track']['artists'][0]['name'])
        playlist.loc[ix, 'artist'] = i['track']['artists'][0]['name']
        playlist.loc[ix, 'name'] = i['track']['name']
        playlist.loc[ix, 'id'] = i['track']['id'] # ['uri'].split(':')[2]
        playlist.loc[ix, 'url'] = i['track']['album']['images'][1]['url']
        playlist.loc[ix, 'date_added'] = i['added_at']

    playlist['date_added'] = pd.to_datetime(playlist['date_added'])  
    
    playlist = playlist[playlist['id'].isin(df['id'].values)].sort_values('date_added',ascending = False)
    
    return playlist
```

```python
from skimage import io
import matplotlib.pyplot as plt

def visualize_songs(df):
    """ 
    Visualize cover art of the songs in the inputted dataframe

    Parameters: 
        df (pandas dataframe): Playlist Dataframe
    """
    
    temp = df['url'].values
    plt.figure(figsize=(15,int(0.5 * len(temp))))
    columns = 5
    
    for i, url in enumerate(temp):
        plt.subplot(len(temp) / columns + 1, columns, i + 1)

        image = io.imread(url)
        plt.imshow(image)
        plt.xticks(color = 'w', fontsize = 0.1)
        plt.yticks(color = 'w', fontsize = 0.1)
        plt.xlabel(df['name'].values[i], fontsize = 12)
        plt.tight_layout(h_pad=0.4, w_pad=0)
        plt.subplots_adjust(wspace=None, hspace=None)

    plt.show()
    
```

# SPotify Playlist to playlist Vector

```python
def generate_plvector(playlist,complete_feature_set,weight_factor):
    
    '''
     playlist:  playlist created by create_necessary_output() using playlist name and id dictionary and original dataset
     complete_feature_set: feature set of original dataset
     
     return 1: single vector of consisting of all the features 
     return 2: complete feature set with playlst songs excluded
     return 3: playlist feature set consisting of songs of playlist dataset found in original dataset
     
    '''
    
    complete_feature_set_pl=complete_feature_set[complete_feature_set['id'].isin(playlist['id'].values)]
    complete_feature_set_pl=complete_feature_set_pl.merge(playlist[['id','date_added']],how='inner',on='id')
    
    complete_feature_set_nonpl=complete_feature_set[~complete_feature_set['id'].isin(playlist['id'].values)]
    playlist_feature_set=complete_feature_set_pl.sort_values('date_added',ascending=False)
    most_recent_date=playlist_feature_set.iloc[0,-1]
    
    
    for index,row in playlist_feature_set.iterrows():
        playlist_feature_set.loc[index,'months_from_recent']=int((most_recent_date.to_pydatetime()-row.iloc[-1].to_pydatetime()).days/30)
        
    playlist_feature_set['weight'] = playlist_feature_set['months_from_recent'].apply(lambda x: weight_factor ** (-x))
    playlist_feature_set_weighted = playlist_feature_set
    
    #                             adding weights to all the values of the playlist data frame
    playlist_feature_set_weighted.update(playlist_feature_set_weighted.iloc[:,:-4].mul(playlist_feature_set_weighted.weight,0))
    playlist_final_feature_set=playlist_feature_set_weighted.iloc[:,:-4]
    playlist_final_feature_set_with_id=playlist_final_feature_set.copy()
    playlist_final_feature_set_with_id['id']=playlist_feature_set_weighted['id']
    
    return playlist_final_feature_set.sum(axis=0),complete_feature_set_nonpl,playlist_final_feature_set_with_id
```

# Cosine Similarity 

Cosine Similarity of the playlist vector with complete feature set

```python
def recommend_cs(df, pl_vector, feature_set):

    #create a dataset out of og dataset that doesn't contain any songs from the playlist
    non_pl_dataset=df[df['id'].isin(feature_set['id'])]
    
    #create a similarity feature using the non-playlist feature set and playlist vector inside the above dataset
    non_pl_dataset['similar']=cosine_similarity(feature_set.drop('id',axis=1),pl_vector.values.reshape(1,-1))[:,0]
    
    #recommend top 20 songs similar to the playlist vector
    non_pl_dataset_top20=non_pl_dataset.sort_values('similar',ascending=False).head(20)

    #gather image url for those 20 songs from the spotify api
    non_pl_dataset_top20['url']=non_pl_dataset_top20['id'].apply(lambda x: sp_dev.track(x)['album']['images'][1]['url'])
    
    return non_pl_dataset_top20
    
    
```

Cosine similarity of User's library with the feature-set consisting of songs by artists that user is following.

```python
def recommend_cs_playlist_artists(df, pl_dataset, feature_set):
    
    '''
       df: original dataset
       pl_dataset: user library dataset
       
    '''
    
    #create a dataset out of og dataset that doesn't contain any songs from the playlist
    non_pl_dataset=df[df['id'].isin(feature_set['id'])]
    
    #filter out the tracks based on same artists as given playlist
    artists=df['artists'][df['id'].isin(pl_dataset['id'])]
    non_pl_dataset_same_artists=non_pl_dataset[non_pl_dataset['artists'].isin(artists)]
    
    # filter feature set wrt to artists
    feature_set=feature_set[feature_set['id'].isin(non_pl_dataset_same_artists['id'])]

    
    #now convert the dataset in a vector
    pl_vector=pl_dataset.drop('id',axis=1).sum(axis=0) # 1 to drop the columns, 0 to sum them index by index
    
    #create a similarity feature using the non-playlist feature set and playlist vector inside the above dataset
    non_pl_dataset_same_artists['similar']=cosine_similarity(feature_set.drop('id',axis=1),pl_vector.values.reshape(1,-1))[:,0]
    
    
    
    #recommend top 20 songs similar to the playlist vector
    non_pl_dataset_top20=non_pl_dataset_same_artists.sort_values('similar',ascending=False).head(20)

    #gather image url for those 20 songs from the spotify api
    non_pl_dataset_top20['url']=non_pl_dataset_top20['id'].apply(lambda x: sp_dev.track(x)['album']['images'][1]['url'])
    
    return non_pl_dataset_top20
```

```python
def create_library_tracks_dataset(df,sp_user,no_of_songs):
    
    '''
       creates a playlistdataset of current user's library tracks that are also present in original dataset
    '''
    read_user_tracks=sp_user.current_user_saved_tracks(limit=no_of_songs, offset=0)
    
    playlist = pd.DataFrame()
    
    for ix, i in enumerate(read_user_tracks['items']):
        #print(i['track']['artists'][0]['name'])
        playlist.loc[ix, 'artist'] = i['track']['artists'][0]['name']
        playlist.loc[ix, 'name'] = i['track']['name']
        playlist.loc[ix, 'id'] = i['track']['id'] # ['uri'].split(':')[2]
        playlist.loc[ix, 'url'] = i['track']['album']['images'][1]['url']
    
    playlist = playlist[playlist['id'].isin(df['id'].values)]
    
    return playlist
    
```

```python
def generate_userlibvector(playlist,complete_feature_set,weight_factor):
    
    '''
     playlist:  playlist created by create_necessary_output() using playlist name and id dictionary and original dataset
     complete_feature_set: feature set of original dataset
     
     return 1: single vector of consisting of all the features 
     return 2: complete feature set with playlst songs excluded
     return 3: playlist feature set consisting of songs of playlist dataset found in original dataset
     
    '''
    
    complete_feature_set_pl=complete_feature_set[complete_feature_set['id'].isin(playlist['id'].values)]
    
    
    complete_feature_set_nonpl=complete_feature_set[~complete_feature_set['id'].isin(playlist['id'].values)]
    playlist_feature_vector=complete_feature_set_nonpl.drop('id',axis=1).sum(axis=0)
    
    
    
    return playlist_feature_vector,complete_feature_set_nonpl,complete_feature_set_pl
```
